#**General Overview of Pandas**

Pandas is a popular open-source data manipulation and analysis library for the Python programming language. It provides data structures for efficiently storing and manipulating large datasets and tools for working with structured data. The primary data structures in Pandas are:

**1. DataFrame:** A two-dimensional table, similar to a spreadsheet or SQL table, where data is organized in rows and columns. Each column can have a different data type (e.g., integers, floats, strings).

**2. Series:** A one-dimensional labeled array that can hold any data type. It is essentially a single column of a DataFrame.

Pandas is widely used in data science, machine learning, and other fields for tasks such as:

1. Data Wrangling: Load, clean, and manipulate data from various sources like CSV files, databases, or APIs.
2. Data Analysis: Perform calculations, statistics, and grouping operations on  data.
3. Data Visualization: Create charts and graphs to explore and understand data visually.
4. Preparation for Machine Learning: Convert data into a format suitable for machine learning algorithms.

**Grouping and Reshaping Data in Pandas**

Data can be grouped and reshaped using groupby() and pivot_table()

Understanding how to group and reshape data is crucial for performing effective exploratory data analysis and generating meaningful insights from the datasets

We will start looking at grouping and reshaping data with pivot tables

The pivot_table() method is used to create pivot tables.

**Parameters in pivot_table() method:**
1. values: The column to aggregate.
2. index: Columns to be used as rows in the pivot table.
3. columns: Columns to be used as columns in the pivot table.
4. aggfunc: Aggregation function (e.g., 'mean', 'sum').

```python
# Create a pivot table to show the average age for each passenger class ('Pclass') and gender ('Sex')
pivot_table = titanic.pivot_table(values='Age', index='Pclass', columns='Sex', aggfunc='mean')

# Print the table
pivot_table
```

```python
# Create a pivot table to analyze the mean survival rate based on gender ('Sex') and passenger class ('Pclass')
# The values to aggregate are the "Survived" column
# The resulting pivot table, pivot_table1, will show the mean survival rate for each combination of gender and passenger class
pivot_table1 = pd.pivot_table(
    titanic,
    values="Survived",
    index="Sex",
    columns="Pclass",
    aggfunc="mean",
)

# Print the table
pivot_table1
```

```python
# Create Pivot Table Three to analyze average fare and median age based on embarkation point ('Embarked')
# The values to aggregate are from the "Fare" and "Age" columns
# The resulting pivot table, pivot_table3, will show the mean fare and median age for each embarkation point
pivot_table3 = pd.pivot_table(
    titanic,
    values=["Fare", "Age"],
    index="Embarked",
    aggfunc={"Fare": "mean", "Age": "median"},
)

# Print table
pivot_table3
```

**Merging Data in Pandas**

In the context of Pandas, the terms "merge" and "join" are often used interchangeably. The merge() function in Pandas provides functionality for combining datasets based on specified columns, similar to the SQL JOIN operation.

There are different merging methods:
**1. Inner Join**
Combining DataFrames based on common columns.
An inner join returns only the rows where there is a match in both DataFrames based on the specified common column. Rows without a match in either DataFrame are excluded from the result.

**2. Left Join**
Combining DataFrames with a left join on a common column.
A left join returns all the rows from the left DataFrame and the matching rows from the right DataFrame. If there is no match in the right DataFrame, the result will contain NaN values for the columns from the right DataFrame.

**3. Right Join**
Combining DataFrames with a right join on a common column.
A right join returns all the rows from the right DataFrame and the matching rows from the left DataFrame. If there is no match in the left DataFrame, the result will contain NaN values for the columns from the left DataFrame.

**4. Outer Join**
Combining DataFrames with an outer join on a common column.
 An outer join returns all the rows from both DataFrames. If there is no match in one of the DataFrames, the result will contain NaN values for the columns from the DataFrame without a match.

 To show how this is done, I will create a sample data that is not related to titanic so that you can visually see whats happening.

```python
# Sample data for merging
student = pd.DataFrame({'key2': ['K0', 'K0', 'K1', 'K2'],
                     'student_id': ['ikeja', 'ikd', 'abk', 'uyo'],
                        'student': ['A0', 'A1', 'A2', 'A3'],
                        'teacher': ['B0', 'B1', 'B2', 'B3']})
student_grade = pd.DataFrame({'student_id': ['ikd', 'uyo', 'lagos', 'benin'],
                               'action': ['K0', 'K0', 'K0', 'K0'],
                                  'grade': ['C0', 'C1', 'C2', 'C3'],
                                  'class': ['D0', 'D1', 'D2', 'D3']})
```

```python
# Merge by inner join
pd.merge(student, student_grade, how= 'inner', on='student_id')
```

```python
# Merge by outer join
pd.merge(student, student_grade, how= 'outer', on='student_id')
```

```python
# Merge by left join
pd.merge(student, student_grade, how= 'left', on='student_id')
```

```python
# Merge by right join
pd.merge(student, student_grade, how= 'right', on='student_id')
```

DataFrames can also be merged based on their indices by using join( method
The join method use (lsuffix and rsuffix parameters) to differentiate columns from left and right

```python
# Join student and student_grade DataFrames
student.join(student_grade, lsuffix='student_id', rsuffix='student_id')
```

**Concetenation**
concatenation is the process of combining two or more DataFrames along a particular axis.

The concat() function is commonly used for this purpose.

We can concetenate along the rows and columns by specifying the axis

1. When concenating the rows, axis=0
2. when concetenating the columns, axis = 1

```python
# Sample data for concetenating
student = pd.DataFrame({'key2': ['K0', 'K0', 'K1', 'K2'],
                     'student_id': ['ikeja', 'ikd', 'abk', 'uyo'],
                        'student': ['A0', 'A1', 'A2', 'A3'],
                        'teacher': ['B0', 'B1', 'B2', 'B3']})
student_grade = pd.DataFrame({'student_id': ['ikd', 'uyo', 'lagos', 'benin'],
                               'action': ['K0', 'K0', 'K0', 'K0'],
                                  'grade': ['C0', 'C1', 'C2', 'C3'],
                                  'class': ['D0', 'D1', 'D2', 'D3']})
```

```python
# concatenate  by columns
pd.concat([student,student_grade],axis=1)
```

```python
# concatenate by rows
pd.concat([student,student_grade],axis=0)
```
