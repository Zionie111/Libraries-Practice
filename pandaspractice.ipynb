There are missing values in Age, Cabin and Embarked.To fill them, the following comands will be used.


```python
# Fill missing values in 'Age' with the median age of passengers
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
```

```python
# Create a new column of Age_Category
bins  = [0,5, 18, 35, 60, 100]
labels = ["Infants", "Children", "Young Adults", "Adults", "Seniors"]
titanic["Age_Category"] = pd.cut(titanic["Age"], bins=bins, labels=labels)
```

```python
titanic["Age"].value_counts()
```

```python
# see the new column
titanic.info()
```

```python
titanic["Age_Category"].unique()
```

```python
# Fill missing Embarked values with the most frequent port
most_frequent_port = titanic["Embarked"].mode()[0]
titanic["Embarked"].fillna(most_frequent_port, inplace=True)
```

```python
titanic["Embarked"].unique()
```

```python
# Create a new category for missing Cabin values
titanic["Cabin"].fillna("Unknown", inplace=True)
```

Creating Plots in Pandas
Different kinds of plots can be created and customized in Pandas. I will show a few plots based on Titanic Dataset

```python
# Histogram of Age Distribution
# Create a single plot
fig, ax = plt.subplots(figsize=(6, 4))

# Plot the histogram of Age
titanic["Age"].plot(kind="hist", ax=ax, title="Age Distribution")
ax.set_xlabel("Age")
ax.set_ylabel("Number of Passengers")

# Display the plot
plt.show()
```

```python
# Create a Bar Chart of Survival Based on Age category
fig, ax = plt.subplots(figsize=(6, 4))

# Create the Bar Chart
titanic.groupby("Age_Category")["Survived"].mean().plot(kind="bar")
plt.title("Age-Based Survival Rates")
plt.xlabel("Age Category")
plt.ylabel("Survival Rate")

plt.show()
```

```python
# BarChart for survival by sex
fig, ax = plt.subplots(figsize=(4, 3))

# Create the bar chart
titanic.groupby("Sex")["Survived"].mean().plot(kind="bar", ax=ax)
ax.set_title("Sex-Based Survival Rates")
ax.set_xlabel("Sex")
ax.set_ylabel("Survival Rate")

# Display the plot
plt.show()
```

```python
# Customized Stacked chart for survival based on Pclass

ax = titanic.groupby("Pclass")["Survived"].value_counts().unstack().plot(
    kind="bar", stacked=True,  color = ["green", "red"]
)

# Add labels directly above each bar segment
for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.0f}",
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha="center",
        va="center",
        fontsize=10,
        color = "white",
    )

# Set title and labels
plt.title("Survival and Passenger Counts by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# Grouping and Reshaping the data using Pivot Tables
# Create the pivot table
pivot_table1 = pd.pivot_table(
    titanic,
    values="Survived",
    index="Sex",
    columns="Pclass",
    aggfunc="mean",
)

pivot_table1
```

Combine data, Join and Merge data


```python
# joining and merging
student = pd.DataFrame({'key2': ['K0', 'K0', 'K1', 'K2'],
                     'student_id': ['ikeja', 'ikd', 'abk', 'uyo'],
                        'student': ['A0', 'A1', 'A2', 'A3'],
                        'teacher': ['B0', 'B1', 'B2', 'B3']})
student_grade = pd.DataFrame({'student_id': ['ikd', 'uyo', 'lagos', 'benin'],
                               'action': ['K0', 'K0', 'K0', 'K0'],
                                  'grade': ['C0', 'C1', 'C2', 'C3'],
                                  'class': ['D0', 'D1', 'D2', 'D3']})
```

```python
# Merge by inner join
pd.merge(student, student_grade, how= 'inner', on='student_id')
```

```python
# Merge by outer join
pd.merge(student, student_grade, how= 'outer', on='student_id')
```

```python
# Merge by left join
pd.merge(student, student_grade, how= 'left', on='student_id')
```

```python
# Join student and student_grade DataFrames
student.join(student_grade, lsuffix='student_id', rsuffix='student_id')
```

```python
# concatenate columns
# works like union
pd.concat([student,student_grade],axis=1)
```

```python
# concatenate rows
# works like union
pd.concat([student,student_grade],axis=0)
```
